## Распознавание именованных сущностей (NER)

Используем `spaCy` для аннотации именованных сущностей

4. Выведем на экран все именованные сущности

````python
import spacy

# Загрузка модели spaCy
nlp = spacy.load("en_core_web_sm")

# Аннотирование предложений
for sentence in sentences:
    doc = nlp(sentence)
    print([(ent.text, ent.label_) for ent in doc.ents])
````

Для анализа тональности используем `TextBlob`

6. Произведем анализ тональности и выведем результаты на экран

````python
from textblob import TextBlob

# Анализ тональности
for sentence in sentences:
    blob = TextBlob(sentence)
    print(f"Предложение: {sentence}")
    print(f"Настроение: {blob.sentiment}")
````
